# Helper Functions

```python
def cv_fit(model, X_train, y_train):

    sns.set(rc={'figure.figsize':(15.7,5)})
    
    tscv = TimeSeriesSplit(n_splits = 5)
    splitted_cv = list(tscv.split(x_all))
        
    rmse = []
    mape = []

    for train_index, test_index in splitted_cv:
        cv_train_X, cv_test_X = X_train.iloc[train_index], X_train.iloc[test_index]
        cv_train_y, cv_test_y = y_train.iloc[train_index], y_train.iloc[test_index]

        model.fit(cv_train_X, cv_train_y)

        preds = model.predict(cv_test_X)
        preds_df = pd.DataFrame({'preds': preds})
        preds_df.index = cv_test_y.index

        rmse_current = round(mean_squared_error(cv_test_y, preds, squared = False), 2)
        mape_current = round(mean_absolute_percentage_error(cv_test_y, preds) * 100, 2)

        rmse.append(rmse_current)
        mape.append(mape_current)
        
        plt.plot(preds_df, label = 'preds');
        plt.plot(cv_test_y, label = 'actual');
        plt.legend();
        plt.show();
        
        print(f"RMSE: {rmse_current}")
        print(f"MAPE: {mape_current}")

    print(f'Average RMSE: {np.mean(rmse)}')
    print(f'Average MAPE: {np.mean(mape)}')
    
    return (np.mean(rmse), np.mean(mape))
```

# Load Data

```python
process = "b21"

PROCESSED_DATA_PATH = f"../outputs/{process}_processed_df.csv"
TRAIN_DATA_PATH = f"../outputs/{process}_train_df.csv"
TEST_DATA_PATH = f"../outputs/{process}_test_df.csv"
TAGS_DICT_PATH = f"../outputs/{process}_tags_dict.json"

all_df = pd.read_csv(PROCESSED_DATA_PATH).set_index('Date')
all_df.index = pd.to_datetime(all_df.index)

train_df = pd.read_csv(TRAIN_DATA_PATH).set_index('Date')
train_df.index = pd.to_datetime(train_df.index)

tags_dict = read_json(TAGS_DICT_PATH)
```

# Split Data

```python
ctrl_tags = tags_dict['controllable']
indct_tags = tags_dict['noncontrollable']

mod_ctrl_tags = [tag for tag in ctrl_tags if tag in all_df.columns]
mod_indct_tags = [tag for tag in indct_tags if tag in all_df.columns]

# response tag after EDA
response = 'RESPONSE'

# function to split x and y
def split_xy(df):
    x_df = df.loc[:, [*mod_ctrl_tags, *mod_indct_tags]]
    y_df = df.loc[:, response]
    
    return x_df, y_df

# split time splits df
x_all, y_all = split_xy(all_df)
x_train, y_train = split_xy(train_df)
```

# Hyperopt

```python
from hyperopt import fmin, tpe, hp, Trials
from sklearn.metrics import make_scorer
from sklearn.model_selection import cross_val_score
```

```python
space = {
    'learning_rate': hp.quniform('learning_rate', 0.05, 0.3, 0.05),
    'max_depth': hp.choice('max_depth', range(3, 5)),
    'min_child_weight': hp.quniform('min_child_weight', 0.5, 3.0, 0.5),
    'lambda': hp.quniform('lambda', 0.5, 5, 0.5),
    'alpha': hp.quniform('alpha', 0.5, 5, 0.5),
    'n_estimators': hp.choice('n_estimators', range(50, 100))
}

tscv = TimeSeriesSplit(n_splits=5)

def objective(params):
    clf = XGBRegressor(**params)
    mape_scores = []
    
    for train_index, test_index in tscv.split(x_all):
        X_train, X_test = x_all.iloc[train_index], x_all.iloc[test_index]
        y_train, y_test = y_all.iloc[train_index].values.ravel(), y_all.iloc[test_index].values.ravel()
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
        mape_scores.append(mape)
        
    score = np.mean(mape_scores)
    return {'loss': score, 'status': 'ok'}

trials = Trials()
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=150,
            trials=trials)

print(best)
```
